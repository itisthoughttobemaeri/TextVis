in questo caso la punteggiatura nello specifico ricerco la:' Virgola PuntoEVirgola DuePunti'

```python
import re
def exp (text):
    pattern = re.compile('(,|;|:|\\.)')
    m=re.findall(pattern, text) 
    return m
```

creato la spirale archimedea usando la equazione r(o)=a+b(o)
t e' angono calcolato in radianti
x e y sono le cordinate del punto dove scrivere il carattere
index Ã¨ indice del array di Stutti


```python
#SPIRALE
import matplotlib.pyplot as plt
import math
a=0.2
b=0.6
index=0
#t=math.radians(len(sTutti)-1)
#lim=((a-b*t)*math.cos(t))
plt.axis([-10,10,-10,10])
l=len(sTutti)
sf=100
for i in range(l+sf,sf,-1):
    t=math.radians(i*5)
    x=(a-b*t)*math.cos(t)
    y=(a-b*t)*math.sin(t)
    plt.text(x,y,sTutti[index])
    index+=1
plt.axis('off')
ax2 = plt.axes([.3, .3, .4, .4])
ax2.imshow(img)
ax2.axis('off')
plt.show()
```
